```python
import random
from matplotlib import pyplot
import matplotlib.pyplot as plt 
```

```python
#programa que usa el metodo monte Carlo para simular una caminata aleatoria 
#Noviembre 3, 2021

def movenX(): #funcion que crea un numero aleatorio para el mov en X 
  deltaX= random.randint(-1000,1000)
  return deltaX/1000

def movenY(): #funcion que crea un numero aleatorio para el mov en Y
  deltaY= random.randint(-1000,1000)
  return deltaY/1000
#imprimir el numero aleatorio 
print(movenX(), "", movenY())

#creacion de vectores que nos guarden cada uno de los pasos 
desplazamientoX=[]
desplazamientoY=[]

Xinicial=0
Yinicial=0
desplazamientoX.append(Xinicial)
desplazamientoY.append(Yinicial)
x=Xinicial
y=Yinicial


NumerodeIteraciones= int(input("¿Cuántas iteraciones o pasos deseas hacer? "))
counter = 0

while counter< NumerodeIteraciones :
  x+=movenX()
  y+=movenY()
  desplazamientoX.append(x)
  desplazamientoY.append(y)
  counter+=1
#imprimir coordenadas finales de la caminata
print("Las coordenadas finales son:")
print(round(x,5),"",round(y,5))

#grafica de la trayectoria aleatoria 
plt.plot(desplazamientoX, desplazamientoY, "#D593EC")
plt.plot(x,y,marker="*",color="m", label="Punto final")
plt.plot(Xinicial,Yinicial,marker="*",color="r",label="Punto inicial")
plt.show()
```

```python
#caminada para N particulas 
#Noviembre 4, 2021

def moverenX(): #funcion que crea un numero aleatorio para el mov en X
  deltaX= random.randint(-1000,1000)
  return deltaX/1000

def moverenY(): #funcion que crea un numero aleatorio para el mov en Y
  deltaY= random.randint(-1000,1000)
  return deltaY/1000

print(movenX(), "", movenY())

NumerodeIteraciones= int(input("¿Cuántas iteraciones o pasos deseas hacer? "))
NumerodeParticulas= int(input("¿Cuántas particulas deseas tener? "))
parts = 0


while parts < NumerodeParticulas: #este while me crea a la particula 
  #creacion de vectores vacios
  desplazamientoX=[]
  desplazamientoY=[]
  Xinicial=0
  Yinicial=0
  desplazamientoX.append(Xinicial)
  desplazamientoY.append(Yinicial)
  x=Xinicial
  y=Yinicial
  counter = 0 
  while counter<NumerodeIteraciones: #y este while ejecuta la caminata de las N particulas 
    x+=movenX()
    y+=movenY()
    desplazamientoX.append(x)
    desplazamientoY.append(y)
    counter+=1
  parts+=1
  plt.plot(desplazamientoX, desplazamientoY)
  plt.plot(Xinicial,Yinicial,marker="*",label="Punto inicial")
  plt.plot(x,y,marker="o",label="Punto final1l")

plt.show()
```
