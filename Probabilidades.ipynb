```python
#Programa que te dice si quedaste en quiebra o el dinero que ganaste u perdiste 
#Georgina Wong ovando 
#Nobiembre 8, 2021
#definir funcion 
def rollDice():
  roll = random.randint(1, 100)

  if roll ==100:
    return False
  elif roll<=50: 
    return False 
  elif 100> roll >50: 
    return True 

#inicializar variables 
NumerodeApuestas= int(input("¿Cuántas veces quieres apostar? "))
DineroparaApuestas= int(input("¿Cuanto dinero quieres apostar? "))
saldo=DineroparaApuestas
Perdidas=0
Ganadas=0

for i in range(0,NumerodeApuestas):
  if rollDice()==True:
    saldo+=1
    Ganadas+=1
  else:
    saldo-=1
    Perdidas+=1
  if saldo==0:
    break
 
if saldo>DineroparaApuestas:
   print("Felicidades!, ganaste", saldo-DineroparaApuestas,"pesos")
   print("Tu dinero actual es de", saldo)
elif DineroparaApuestas>saldo:
  print("Lo sentimos has perdido", DineroparaApuestas-saldo, "pesos")
  print("Tu dinero actual es de", saldo)
elif saldo==0:
   print("¡ESTAS EN QUIEBRAAAA!, ya no tienes saldo suficiente para seguir jugando :(")
```
