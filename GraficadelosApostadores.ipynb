```python
import random 
from matplotlib import pyplot
import matplotlib.pyplot as plt 
import numpy as np
```

```python
#Programa que te grafica el juego de apuesta de cada jugador, apostando el doble cada que gana 

#definir funcion 
def rollDice():
  roll=random.randint(1,100)

  if roll==100:
    return False 
  elif roll <=50:
    return False
  elif 100> roll > 50:
    return True

#crear arreglos que guardaran nuestros datos
counter=0
grafica1=[]
grafica2=[]
datosJuegos=[]
datosSaldo=[]

#inicializar variables 
numeroDeJugadores= int(input("¿Cuántos jugadores desean jugar? "))
NumJuegos= int(input("¿Cuántas veces deseas jugar? "))
saldo_0= int(input("¿Cuánto dinero dispone para jugar? "))

while counter<numeroDeJugadores:
  numeroDeJuegos= NumJuegos 
  saldoInicial= saldo_0
  datosJuegos.append(numeroDeJuegos)
  datosSaldo.append(saldoInicial)
  counter=counter+1

counter2=0
multiplicador=0
apuesta0=100
apuesta1=apuesta0
while counter2<numeroDeJugadores:
  grafica2.append(datosSaldo[counter2])
  grafica1.append(0)

  saldo=datosSaldo[counter2]

  for i in range(0,datosJuegos[counter2]):
    if rollDice()==False:
      saldo-=apuesta1
      apuesta1=apuesta0
      multiplicador=0
    else:
      saldo+=apuesta1
      multiplicador+=1
      apuesta1=apuesta0*(2**multiplicador)
    if saldo==0:
      break
    grafica2.append(saldo)
    grafica1.append(i)

  counter2=counter2+1

  plt.plot(grafica1, grafica2)
  plt.plot(grafica1[0],grafica2[0],marker=".",label="Inicio")
  plt.show
  grafica1=[]
  grafica2=[]
```

```python

def rolldice():
  roll = random.randint(1,100)
  
  if roll == 100:
    return False
  elif roll <= 50:
    return False
  elif 100> roll > 50:
    return True

saldoinicial=10000
saldo=saldoinicial
jugadasmax =100
apuesta=100
jugadores=100
parts=0
ganancias=[]
K=np.arange(1,jugadores+1,1)
T=np.arange(0,jugadasmax+1,1)


while parts<jugadores:
  ganancias.append(saldo)
  saldo=saldoinicial

  for i in K:
    if rolldice()==True:
      saldo=saldo+apuesta
      apuesta+=apuesta*2
    else:
      saldo=saldo-apuesta
      apuesta=100
    if saldo<=0:
      saldo=0
      apuesta=0
       
    ganancias.append(saldo)
  parts +=1
    
  plt.plot(T,ganancias)
  ganancias=[]
  
plt.show()

print(ganancias)    


```
