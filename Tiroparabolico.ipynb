```python
#Programa que calcula la posición, velocidad y aceleración del tiro parabolico
#autor: Georgina Wong
#fecha: Septiembre 29, 2021

import matplotlib.pyplot as plt 
import numpy as np 

#Definir funciones
def function1(t): #posición en x
  return x0 +vx0*t

def function2(t): #velocidad en x
  return (function1(t+h)-function1(t))/h

def function3(t): #aceleracion en x 
  return (function2(t+h)-function2(t))/h

def function4(t): #posicion en y
  return y0 + vy0*t -(1/2)*(9.81)*(t**2)

def function5(t): #velocidad en y 
  return (function4(t+h)-function4(t))/h

def function6(t): #aceleración en y 
  return (function5(t+h)-function5(t))/h


tiempo= np.arange(0,80, 0.01)

#Pedir variables
print("Ingresa los datos para calcular las derivadas númericas de la función")
print("¿Cual es la posición inicial en x y en y? [m]:")
x0= float (input("x_0="))
y0= float (input("y_0="))
print("¿Cual es la velocidad inicial en x y en y? [m/s]:")
vx0= float (input("vx_0=" ))
vy0= float (input("vy_0=" ))
tpunto= float (input("Digite el tiempo [s]: "))
h= float (input("Intervalo de tiempo para calcular la derivada: "))

#impresión de resultados 
print("La posición del objeto en el eje x, en el tiempo", tpunto ," segundos es: ", function1(tpunto), "m")
print("La posición del objeto en el eje y, en el tiempo", tpunto ," segundos es: ", function4(tpunto), "m")
print("La velocidad en x, en el tiempo", tpunto ," segundos es: ", function2(tpunto), "m/s")
print("La velocidad en y, en el tiempo", tpunto ," segundos es: ", function5(tpunto), "m/s")
print("La aceleración en x, en el tiempo", tpunto ," segundos es: ", function3(tpunto) , "m/s^2")
print("La aceleración en y, en el tiempo", tpunto ," segundos es: ", function6(tpunto) , "m/s^2")
print("La magnitud del vector de posición es", np.sqrt(function1(tpunto)**2+function4(tpunto)**2))
print("La magnitud del vector de velocidad es", np.sqrt(function2(tpunto)**2+function5(tpunto)**2))
print("La magnitud del vector de aceleración es", np.sqrt(function3(tpunto)**2+function6(tpunto)**2))

#Gráficas
plt.plot(tiempo, function1(tiempo), label="r_x(t)")
plt.plot(tiempo, function2(tiempo), label="v_x(t)")
plt.plot(tiempo, function3(tiempo), label="a_x(t)")
plt.plot(tiempo, function4(tiempo), label="r_y(t)")
plt.plot(tiempo, function5(tiempo), label="v_y(t)")
plt.plot(tiempo, function6(tiempo), label="a_y(t)")
plt.xlabel ("Variable Independiente  (t) [s]")
plt.ylabel ("f(t)")
plt.legend()
plt.show()
```
