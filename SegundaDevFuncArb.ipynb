```python
#Programa que calcula la primera y segunda derivada de una función 
#autor: Georgina Wong
#fecha: Septiembre 27, 2021
#mejorar este código 

import matplotlib.pyplot as plt 
import numpy as np 

#Definir funciones
def function1(t):
  return t**3+cte*t**2+masa*t
def function2(t):
  return (function1(t+h)-function1(t))/h
def function3(t):
  return (function2(t+h)-function2(t))/h

tiempo= np.arange(1,3, 0.01)

#Pedir variables
print("Ingresa los datos para calcular las derivadas númericas de la función")
cte= float (input("constante del oscilador: "))
masa= float (input("masa del oscilador en kilogramos: "))
h= float (input("Intervalo de tiempo para calcular la derivada: "))
tpunto= float (input("Punto en el que quieres evaluar a la primer y segunda derivada de la función: "))
#impresión de resultados 
print("La posición en el punto dado es: ", function1(tpunto), "metros")
print("La velocidad en el punto dado es: ",function2(tpunto), "m/s")
print("La aceleración en el punto dado es: ", function3(tpunto), "m/s^2")
#gráfica de las funciones 
plt.plot(tiempo, function1(tiempo), label="F(t)")
plt.plot(tiempo, function2(tiempo), label="derivada de F(t)")
plt.plot(tiempo, function3(tiempo), label="Segunda derivada de F(t)")
plt.xlabel ("Variable Independiente  (t) [units]")
plt.ylabel ("f(t)   [units]")
plt.legend()
plt.show()
```
